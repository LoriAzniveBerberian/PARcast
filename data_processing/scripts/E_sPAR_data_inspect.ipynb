```python
# import the libraries we need
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the data
df = pd.read_csv('/Users/loriberberian/Desktop/bioSCape/PARcast/data_processing/data/raw/E_sPAR/E_sPAR_20260402_0002.CSV',
                 parse_dates=['datetime'])

# quick look at the data
df.head(10)
```

```python
#summary statistics
df.describe()
```

```python
# Plot PPFD over time
plt.figure(figsize=(14, 5))
plt.plot(df['datetime'], df['ppfd_umol_m2_s'], color='steelblue', linewidth=1)
plt.xlabel('Time')
plt.ylabel('PPFD (\u00b5mol/m\u00b2/s)')
plt.title('PARcast E_s(PAR) Reference \u2014 E_sPAR_20260402_0002')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# filter to only show outdoor readings (PPFD > 10)
df_outdoor = df[df['ppfd_umol_m2_s'] > 10]

plt.figure(figsize=(14, 5))
plt.plot(df_outdoor['datetime'], df_outdoor['ppfd_umol_m2_s'],
         color='steelblue', linewidth=1.5)
plt.xlabel('Time')
plt.ylabel('PPFD (\u00b5mol/m\u00b2/s)')
plt.title('PARcast E_s(PAR) Reference \u2014 Outdoor Readings Only')
plt.xticks(rotation=45)

# as mean line
mean_ppfd = df_outdoor['ppfd_umol_m2_s'].mean()
plt.axhline(y=mean_ppfd, color='red', linestyle='--',
            label=f'Mean: {mean_ppfd:.1f} \u00b5mol/m\u00b2/s')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Summary statistics for outdoor readings only
print("=== PARcast E_s(PAR) Reference Summary ===")
print(f"File: E_sPAR_20260402_0002.CSV")
print(f"Deployment start: {df['datetime'].min()}")
print(f"Deployment end:   {df['datetime'].max()}")
print(f"Total readings:   {len(df)}")
print(f"Outdoor readings: {len(df_outdoor)}")
print(f"Mean PPFD:        {df_outdoor['ppfd_umol_m2_s'].mean():.1f} \u00b5mol/m\u00b2/s")
print(f"Max PPFD:         {df_outdoor['ppfd_umol_m2_s'].max():.1f} \u00b5mol/m\u00b2/s")
print(f"Min PPFD:         {df_outdoor['ppfd_umol_m2_s'].min():.1f} \u00b5mol/m\u00b2/s")
print(f"Std deviation:    {df_outdoor['ppfd_umol_m2_s'].std():.1f} \u00b5mol/m\u00b2/s")
```
